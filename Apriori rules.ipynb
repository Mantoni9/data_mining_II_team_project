```python
import pandas as pd
import numpy as np
from efficient_apriori import apriori
```

```python
# adjust file paths
train_sessions_path = "C:/Users/aminm/OneDrive/Desktop/sessions_train.csv"
test_sessions_path = "C:/Users/aminm/OneDrive/Desktop/sessions_test_task1.csv"
products_path = "C:/Users/aminm/OneDrive/Desktop/products_train.csv"

# load into dataframes
train_df = pd.read_csv(train_sessions_path)
test_df = pd.read_csv(test_sessions_path)
products_df = pd.read_csv(products_path)

# print info to inspect datatypes
print(train_df.info())
print("---------------")
print(test_df.info())
print("---------------")
print(products_df.info())

# show first rows
display(train_df.head())
display(test_df.head())
display(products_df.head())
```

```python
# we can use the python string methods directly on pandas dataframe columns by adding .str in front of the method call
# this way the operation is vectorized and faster than if we would use the apply method and a custom function
train_df['prev_items'] = train_df['prev_items'].str.strip("[']").str.replace("\n", "").str.split("' '")
test_df['prev_items'] = test_df['prev_items'].str.strip("[']").str.replace("\n", "").str.split("' '")
```

```python
#input data as a list of lists
data = train_df["prev_items"].tolist()

# Call the apriori function with the input data and minimum support
itemsets, rules = apriori(data, min_support=0.0001)

# Print the frequent itemsets and association rules
print("Frequent Itemsets:")
for itemset in itemsets:
    print(itemset)

print("\nAssociation Rules:")
for rule in rules:
    print(rule)
```
