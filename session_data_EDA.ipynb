```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# define filepaths
train_sessions_path = "./data/sessions_train.csv"

# load into dataframes
train_df = pd.read_csv(train_sessions_path)
```

```python
# split item strings into list
df = pd.DataFrame(train_df['prev_items'].str.strip("[']").str.replace("\n", "").str.split("' '"))

# Count the length of each list
lengths = df['prev_items'].apply(lambda x: len(x))
lengths.nlargest(50)
```

```python
lengths.describe()
```

```python
# Plot the histogram
plt.hist(lengths, bins=range(1, lengths.max() + 2))
plt.xlabel('List length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.rcParams['figure.dpi'] = 300

# Plot the histogram
plt.hist(lengths, bins=range(1, lengths.max() + 2))
plt.xlabel('List length')
plt.ylabel('Frequency')

plt.title('Frequency of previous items list length')

# Set the x-axis to a logarithmic scale
plt.xscale('log')

plt.show()
```
